```python
# import all dictionaries
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_ray(point: list, angle_degrees: list, length=100):
    for i in range(len(point)):
        #      
        angle_radians = np.radians(angle_degrees[i])

        #    
        end_point1 = [
            point[i][0] + length * np.cos(angle_radians),
            point[i][1] + length * np.sin(angle_radians)
        ]

        end_point2 = [
            point[i][0] - length * np.cos(angle_radians),
            point[i][1] - length * np.sin(angle_radians)
        ]

        #  
        plt.plot([end_point2[0], end_point1[0]], [end_point2[1], end_point1[1]])

#      
points = [(1, 6), (-5, -9), (16, 5), (-14, -21), (10, -2), (21, 7), (2, -7), (-7, -3)]
angles = [42, 36, 158, 360, 38, 25, 1, 248]

#  
length = 100

#  
plt.figure(figsize=(12, 12))

#  
plt.axhline(0, color='gray', linestyle='dashed')

#  
plt.axvline(0, color='gray', linestyle='dashed')

#  
for i in range(30):
    plt.axhline(i, color='gray', linestyle='dotted')
    plt.axvline(i, color='gray', linestyle='dotted')
    plt.axhline(-i, color='gray', linestyle='dotted')
    plt.axvline(-i, color='gray', linestyle='dotted')

#     
plot_ray(points, angles, length)

#        
plt.gca().set_aspect('equal', adjustable='box')

x_min, x_max, y_min, y_max = -16, 30, -15, 15
plt.axis([x_min, x_max, y_min, y_max])

#  
plt.legend()

#  
plt.show()
```

```python
def find_location(transmitters, bearings):
  """            .

  Args:
    transmitters:  .
    bearings:      .

  Returns:
     .
  """

  #   .

  if len(transmitters) != len(bearings):
    raise ValueError("      .")

  #  .

  x = 0
  y = 0
  for i in range(len(transmitters)):
    x += transmitters[i][0] * math.cos(bearings[i])
    y += transmitters[i][1] * math.sin(bearings[i])

  #     .

  return (x / len(transmitters), y / len(transmitters))


#  .

transmitters = [(10, 10), (20, 20), (30, 30)]
bearings = [30, 60, 90]

location = find_location(points, angles)

print(location)
```
